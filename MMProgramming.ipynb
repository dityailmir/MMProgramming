```
print("==============================================")
print("R U M U S    B A N G U N   R U A N G")
print("==============================================")
print("R u m u s  T a b u n g")
print("R u m u s  K e r u c u t")
print("R u m u s  B o l a")
print("pilihan: ")
pil = str(input())
while (pil != "0"):
  if (pil == "1"):
    print(" ")
    print("====RUMUS TABUNG===")
    print("Diketahui")
    t = int(input("tinggi : "))
    r = int(input("jari-jari : "))
    phi = 3.14
    print("phi : ", phi)
    v = phi * r * r * t
    l = 2 * phi * r * (r + t)
    print("-----------------------------------------")
    print("volume tabung : ", v)
    print("luas permukaan : ", l)
    print("-----------------------------------------")
  elif (pil == "2"):
    print(" ") 
    print("====RUMUS KERUCUT===")
    print("Diketahui")
    ti = float(input("tinggi : "))
    jari = float(input("jari-jari : "))
    s = int(input("apotema : "))
    pi = 3.14
    print("phi : ", pi)
    vol = 1/3 * pi * jari * jari * ti
    lpk = pi * jari * (jari + s)
    print("-----------------------------------------")
    print("volume kerucut : ", vol)
    print("luas permukaan : ", lpk)
    print("-----------------------------------------")
  elif (pil == "3"):
    print(" ") 
    print("====RUMUS BOLA===")
    print("Diketahui")
    j = float(input("jari-jari : "))
    ph = 3.14
    print("phi : ", ph)
    volume = 4/3 * ph * j * j * j
    lpb = 4 * ph * j * j
    print("-----------------------------------------")
    print("volume bola : ", volume)
    print("luas permukaan : ", lpb)
    print("-----------------------------------------")
  else:
    print("inputan tidak valid")
  print("pilihan: ")
  pil = str(input())

print("============================================")
print("D i t y a  I l m i  R i z q i")
print("     ")
print("1 3 0 3 1 9 1 1 1 9")
print("============================================")
```
